# Example 1:
The Cal Poly course catalog (http://catalog.calpoly.edu/collegesandprograms/collegeofsciencemathematics/statistics/#courseinventory) contains a list of courses offered by the Statistics department. Scrape this website to obtain a DataFrame, where the unit of observation is a course. Use this DataFrame to answer the following questions:



How many 300-level courses does the Statistics department offer?

```python
#Step 1:Importing libraries
import requests
from bs4 import BeautifulSoup
```

```python
#Step 2:Adding URL
url="https://catalog.calpoly.edu/collegesandprograms/collegeofsciencemathematics/statistics/#courseinventory"
```

```python
#Step 3:Creating Soup object
urldata=requests.get(url)
urldata
soup=BeautifulSoup(urldata.text,'html.parser')
soup.prettify()
soup.find_all(class_="courseblocktitle")
```

```python
soup.find(class_="courseblocktitle").findChild("strong")
```

```python
#Step 4: Retrieve the Data
list1=[]
for i in soup.find_all(class_="courseblocktitle"):
  text1=i.text[0:8]
  list1.append(text1)
list1

```

```python
import pandas as pd
df=pd.DataFrame(list1,columns=["Course Code"])
df
```

```python
df2=df[df['Course Code'].str.get(5)=='3']
df2.head()
```

```python
#Step 4: Retrieve the Data
list1=[]
for i in soup.find_all(class_="RatingHeader__StyledClass-sc-1dlkqw1-3 eXfReS"):
  text1=i.text[0:8]
  list1.append(text1)
list1
```
